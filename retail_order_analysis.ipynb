```python
import pandas as pd 
# Remove any outlier values within the ship_mode column, getting rid of any 'Not Available' and 'unknown' data categories
df = pd.read_csv('orders.csv', na_values = ['Not Available', 'unknown'])
```

```python
# Restructure the dataframe's column names to a more technical format(lower case and replace spaces with underscores)
df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
```

```python
df.columns = df.columns.str.lower()
```

```python
# Create new columns by utilizing given columns to perform more detailed quering on SQL

df['discount'] = df['list_price'] * df['discount_percent'] * 0.01
```

```python
df['sale_price'] = df['list_price'] - df['discount']
```

```python
df['profit'] = df['sale_price'] - df['cost_price']
```

```python
#Convert the Order dates into an datetime datatype
df['order_date'] = pd.to_datetime(df['order_date'], format = "%Y-%m-%d")
```

```python
# Drop unnecessary columns out the dataframe
df.drop(columns = ['list_price', 'cost_price', 'discount_percent'], inplace=True)
```
